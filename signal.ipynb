```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import firwin, hamming
```

```python
fs = 32
t = np.arange(0,1,1/fs)
signal = np.sin(2*np.pi*t)
plt.plot(t,signal)
```

```python
noise = np.random.normal(0,0.1,t.size)
signalNoise = signal+noise
plt.plot(t, signalNoise)
```

```python
scaling = 7
signalNoise = signalNoise*(2**scaling)
plt.plot(t,signalNoise)
```

```python
signalRounded = list((np.rint(signalNoise)).astype(int))
signalBinary = []
for i in signalRounded:
    signalBinary.append(np.binary_repr(i, width = 16))
print(signalBinary)
```

```python
f = open('signal.data', 'w')
f.write('\n'.join(signalBinary))
f.close()
```
